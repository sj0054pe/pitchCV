```
#https://sabopy.com/py/matplotlib-animation-5/

import itertools

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation


def data_gen():
    for cnt in itertools.count():
        t = cnt / 10
        yield t, np.sin(2*np.pi*t) * np.exp(-t/10.)


def init():
    ax.set_ylim(-1.1, 1.1)
    ax.set_xlim(0, 10)
    del xdata[:]
    del ydata[:]
    line.set_data(xdata, ydata)
    return line,

def run(data):
    # update the data
    t, y = data
    xdata.append(t)
    ydata.append(y)
    xmin, xmax = ax.get_xlim()

    if t >= xmax:
        ax.set_xlim(xmin, 2*xmax)
        ax.figure.canvas.draw()
    line.set_data(xdata, ydata)

    return line,

fig, ax = plt.subplots()
line, = ax.plot([], [], lw=2)
ax.grid()
xdata, ydata = [], []

ani = animation.FuncAnimation(fig, run, data_gen, interval=10, init_func=init)
plt.show()
```

```
#!/usr/bin/env python
# -*- coding: utf-8 -*-

import math

import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation

import pandas as pd
def read_file():
    a=pd.read_csv("file.csv", encoding = "shift-jis")

def _update(frame, x, y, list_coords_y, OUTPUT_PATH):
    """グラフを更新するための関数"""
    # 現在のグラフを消去する
    #print(frame)
    plt.cla()
    # データを更新 (追加) する
    x.append(frame)

    print(list_coords_y[int(frame)])
    y.append(list_coords_y[int(frame)])
    # 折れ線グラフを再描画する
    plt.plot(x, y)
    plt.xlim((0, len(list_coords_y)))
    plt.ylim((0,1000))
    plt.savefig(OUTPUT_PATH+"/%s.jpg" % (str(frame).zfill(4)), transparent = True, bbox_inches = 'tight', pad_inches = 0.1)
    plt.show()


def main():
    OUTPUT_PATH="./fig"
    # 描画領域
    fig = plt.figure(figsize=(10, 6))
    # 描画するデータ (最初は空っぽ)
    x = []
    y = []

    params = {
        'fig': fig,
        'func': _update,  # グラフを更新する関数
        'fargs': (x, y, list_coords_y, OUTPUT_PATH),  # 関数の引数 (フレーム番号を除く)
        'frames': np.arange(0, len(list_coords_y), 1),  # フレーム番号を生成するイテレータ
        'interval': 10,  # 更新間隔 (ミリ秒)
        'repeat': False,  # 繰り返さない
    }
    #print(params)
    try:
      import os 
      os.mkdir(OUTPUT_PATH)
    except:
      print('the directory already exists')
    anime = animation.FuncAnimation(**params)
    plt.show()
    # グラフを保存する
    anime.save('sin2.gif', writer='pillow')
    print('XXXXXXXXXXXX')


if __name__ == '__main__':
    main()
```
