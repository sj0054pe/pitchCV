```
import os
import cv2

def check_video_info(PATH, FILE):
  if os.path.isfile(PATH+FILE) ==True:
    print("パス「%s」において、\nファイル名「%s」の存在が確認できました。\n" % (PATH, FILE))
  else:
    print("パス「%s」において、\nファイル名「%s」の存在が確認できません。パスを見直してください。" % (PATH, FILE))

  cap = cv2.VideoCapture(PATH+FILE)

  fourcc = cv2.VideoWriter_fourcc(*'XVID')
  width= cap.get(cv2.CAP_PROP_FRAME_WIDTH)
  height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
  fps = cap.get(cv2.CAP_PROP_FPS)

  axes = None
  num_frames = cap.get(cv2.CAP_PROP_FRAME_COUNT)

  print(cap)
  print("height : ", height)
  print("width  : ", width)
  print("fps     : ", int(fps))
  print("frames : " , num_frames)

  return height, width, int(fps), num_frames, fourcc


print("動画の基礎データを取得します... ")
print()
[height, width, fps, num_frames, fourcc]=check_video_info(INPUT_PATH, INPUT_FILE)
```
